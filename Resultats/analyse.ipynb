```python
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore
# Lecture des données depuis le fichier CSV
data = pd.read_csv('temps_execution.csv')

# Extraction des colonnes
nbPointsTotal = data['NbPointsTotal']
tempsListe = data['TempsListe']
tempsHachage = data['TempsHachage']
tempsArbre = data['TempsArbre']

nbPointsTotal = np.array(nbPointsTotal)
tempsListe = np.array(tempsListe)
tempsHachage =np.array(tempsHachage)
tempsArbre=np.array(tempsArbre)
# Création des graphiques
plt.figure(figsize=(10, 5))

# Graphique pour la liste chaînée
plt.plot(nbPointsTotal, tempsListe, label='Liste chaînée', marker='o', color='blue')

# Ajout de titres et légendes
plt.title('Temps de calcul en fonction du nombre de points total')
plt.xlabel('Nombre de points total')
plt.ylabel('Temps de calcul (s)')
plt.legend()

# Affichage du graphique
plt.grid(True)
plt.show()
plt.savefig("im3.png")
```

```python
def clef(x, y):
    return y + (x + y) * (x + y + 1) // 2

def tester_fonction_hachage():
    nb_tests = 0
    nb_collisions = 0
    clés = set() 

    # Test de toutes les paires de coordonnées (x, y) avec x et y allant de 1 à 10
    for x in range(1, 11):
        for y in range(1, 11):
            nb_tests += 1
            c = clef(x, y)
            if c in clés:
                nb_collisions += 1
            else:
                clés.add(c)

    print("Nombre total de tests:", nb_tests)
    print("Nombre de collisions:", nb_collisions)
    print("Pourcentage de collisions:", nb_collisions / nb_tests * 100, "%")

if __name__ == "__main__":
    tester_fonction_hachage()
```
